```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Nate_part.xlsx')
```

```python
#(format for Capex and Opex)
df.dtypes
```

```python
df.shape
```

```python
#objetivo analysis
df['Objetivo'].isnull().sum()
```

```python
#Objetivo barchart 

counts = df['Objetivo'].value_counts()

plt.bar(counts.index, counts.values)

plt.xlabel('Objetivo')
plt.ylabel('Count')
plt.title('Objectivos Chicos')

plt.xticks(rotation=45, ha='right')


plt.show()
```

```python
#Agrupación IT de Ejecución analysis 

df['Agrupación IT de Ejecución'].isnull().sum()
```

```python
#Agrupación IT de Ejecución barchart 

counts = df['Agrupación IT de Ejecución'].value_counts()

plt.bar(counts.index, counts.values)

plt.xlabel('Agrupación IT de Ejecución')
plt.ylabel('Count')
plt.title('Agrupación IT de Ejecución De Mierda')

plt.xticks(rotation=45, ha='right', fontsize=5)


plt.show()
```
