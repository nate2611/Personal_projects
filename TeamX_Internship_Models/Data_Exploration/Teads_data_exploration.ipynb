```python
dates=[]
for d in df['Day']:
    date = d.split('T')[0]
    dates.append(date)

df['date']=dates
df['date']=df['date'].astype('datetime64[ns]')
```

```python
df_2022_2023=pd.read_csv('Mar-2022 - Jun-2023.csv')
dates=[]
for d in df_2022_2023['Day']:
    date = d.split('T')[0]
    dates.append(date)

df_2022_2023['date']=dates
df_2022_2023['date']=df_2022_2023['date'].astype('datetime64[ns]')
df_2022_2023.sort_values(by=['date'])
df_2022_2023.reset_index(drop=True, inplace=True)
```

```python
df_2022_2023.shape
df_2022_2023=df_2022_2023[df_2022_2023['date']>'2022-12-09']
df=pd.concat([df, df_2022_2023], ignore_index=True)
```

```python
phase=[]

for index, row in df.iterrows():

    if 'AWRN' in row['Campaign name']:
        phase.append("Awareness")
    elif 'AWRN' in row['Line item name']:
        phase.append("Awareness")
    elif 'CONS' in row['Campaign name']:
        phase.append("Consideration")
    elif 'CONS' in row['Line item name']:
        phase.append("Consideration")
    elif 'PRCH' in row['Campaign name']:
        phase.append("Decision")
    elif 'PRCH' in row['Line item name']:
        phase.append("Decision")
    else:
        phase.append(None)

df['phase']=phase
```

```python
df.shape
```

```python
df=df.sort_values(by=['date'])
df.reset_index(drop=True, inplace=True)
df.to_csv('concatenated.csv')
```
