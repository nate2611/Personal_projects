```python
#Creation of phase column
#Extracting the phase column
phase=[]

for name in df['Campaign name']:
    if "AWRN" in name :
        phase.append('Consideration')
    elif name == 'Website conversions':
        phase.append('Decision')
    elif name == 'Lead generation':
        phase.append('Decision')
    elif name == 'Video views':
        phase.append('Awareness')
    elif name == 'Brand awareness':
        phase.append('Awareness')
    else:
        phase.append(None)

df['phase']=phase
```
